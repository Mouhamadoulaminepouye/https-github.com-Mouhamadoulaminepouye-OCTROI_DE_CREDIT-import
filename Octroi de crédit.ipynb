### 3. Analyse Univariée des Variables Catégorielles <a id="3-analyse-univariée-des-variables-catégorielles"></a>

L'analyse univariée nous aide à comprendre les caractéristiques individuelles de chaque variable catégorielle, notamment leur distribution.

**Approche :** Pour chaque variable catégorielle, visualisez la distribution de ses catégories.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_categorical_distribution(data, var, order, title=None, xlabel=None):
    """
    Affiche la distribution d'une variable catégorielle avec les comptes et pourcentages,
    et affiche un tableau résumé avec des nuances de bleus du plus petit au plus grand pourcentage.
    
    Args:
    - data (DataFrame): Le DataFrame contenant les données.
    - var (str): Le nom de la variable catégorielle à visualiser.
    - order (list): L'ordre des modalités à afficher.
    - title (str): Le titre du graphique.
    - xlabel (str): Le label pour l'axe des x.
    """
    # Calcul des décomptes et pourcentages
    counts = data[var].value_counts().reindex(order)
    percentages = data[var].value_counts(normalize=True).reindex(order) * 100
    
    # Création du DataFrame résumé
    summary_df = pd.DataFrame({'Comptes': counts, 'Pourcentages': percentages})
    
    # Affichage du tableau résumé avec nuances de bleus
    print(f"Résumé pour {var}:\n")
    cm = sns.light_palette("blue", as_cmap=True)
    display(summary_df.style.background_gradient(cmap=cm).format({'Pourcentages': "{:.2f}%"}))
    
    # Création du graphique
    plt.figure(figsize=(10, 6))
    sns.barplot(x=summary_df.index, y='Comptes', data=summary_df, palette='Set2', order=order)
    plt.title(title if title else f'Distribution de {var}')
    plt.xlabel(xlabel if xlabel else var)
    plt.xticks(rotation=45)
    plt.ylabel('Nombre d\'Observations')
    
    # Ajout des étiquettes de pourcentage sur chaque barre
    for i, p in enumerate(plt.gca().patches):
        label = f'{summary_df["Pourcentages"].iloc[i]:.2f}%'
        plt.annotate(label, (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='center', fontsize=10, color='black', rotation=0, xytext=(0, 8),
                     textcoords='offset points')
    
    plt.show()
```

```python
# Mapping des variables aux ordres spécifiques de leurs catégories
orders = {
    'Age du client': ['moins de 23 ans', 'de 23 à 39 ans', 'de 40 à  50 ans', 'plus de  50 ans'],
    'Situation familiale': ['célibataire', 'marié', 'divorcé', 'veuf'],
    'Ancienneté': ['anc. 1 an  ou moins', 'anc. de 1  à 4 ans', 'anc. de 4  à 6 ans', 'anc. de 6  à 12 ans', 'anc. plus   12 ans'],
    'Domiciliation du salaire': ['Non domicilié', 'Oui domicilié'],
    'Domiciliation de l\'épargne': ['pas d\'épargne', 'moins de 10K épargne', 'de 10 à 100K épargne', 'plus de 100K épargne'],
    'Profession': ['autre', 'employé', 'cadre'],
    'Moyenne encours': ['moins de 2K encours', 'de 2 à 5 K  encours', 'plus de 5 K encours'],
    'Moyenne des mouvements': ['moins 10 K  mouvt', 'de 10 à 30K mouvt', 'de 30 à 50K mouvt', 'plus de 50K mouvt'],
    'Cumul des débits': ['moins de 40  débits', 'de 40 à 100  débits', 'plus de 100  débits'],
    'Autorisation de découvert': ['découvert  interdit', 'découvert  autorisé'],
    'Interdiction de chéquier': ['chéquier   interdit', 'chéquier   autorisé']
}

# Itération sur chaque variable et son ordre spécifique pour la visualisation
for var, order in orders.items():
    plot_categorical_distribution(data, var, order, title=f'Distribution de {var}')
```

<span style="color: red;"><strong>L'analyse de la distribution des variables catégorielles nous a fourni des insights précieux sur notre clientèle :</strong></span>


1.  **Âge du Client** : Nous observons que la banque attire principalement des clients de deux groupes d'âge spécifiques : les jeunes adultes (23 à 39 ans) et les adultes de 40 à 50 ans. Cette répartition peut indiquer des besoins financiers distincts qui pourraient être exploités par des offres ciblées.

### 4. Analyse des Corrélations et Intérêt <a id="4-analyse-des-corrélations-et-intérêt"></a>

1.  **Éviter la Multicollinéarité** : Dans les modèles de régression, la multicollinéarité (corrélation élevée entre variables indépendantes) peut fausser les estimations des coefficients et rendre difficile l'interprétation des résultats. Identifier et traiter la multicollinéarité est donc crucial pour la robustesse du modèle.
2.  **Simplification du Modèle** : En éliminant les variables redondantes (fortement corrélées entre elles), nous pouvons simplifier notre modèle sans perdre de pouvoir prédictif significatif. Cela rend le modèle plus facile à interpréter et souvent plus performant.
3.  **Identification des Relations** : Comprendre comment les variables sont liées entre elles peut fournir des insights précieux pour l'analyse et la prise de décision.


```python
from scipy.stats import chi2_contingency

# définition de la fonction cramers_v
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2_corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    r_corr = r - ((r-1)**2)/(n-1)
    k_corr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))

# Calcul de la matrice des coefficients de Cramér V avec les données ajustées et la fonction redéfinie
cramers_v_matrix_updated = pd.DataFrame(np.zeros((len(cat_vars), len(cat_vars))), index=cat_vars, columns=cat_vars)

for col1 in cat_vars:
    for col2 in cat_vars:
        cramers_v_matrix_updated.loc[col1, col2] = cramers_v(data[col1], data[col2])

# Affichage de la carte de chaleur pour la matrice recalculée
plt.figure(figsize=(12, 10))
sns.heatmap(cramers_v_matrix_updated, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Carte de Chaleur des Coefficients de Cramér V entre Variables Catégorielles (Après Ajustements)')
plt.show()
```

<span style="color: red;"><strong>Observations issues de cette analyse des corrélations des variables Xi :</strong></span>


*   **Variables Clés Identifiées :** "Domiciliation du salaire" et "Moyenne encours" se distinguent par leurs corrélations relativement plus élevées avec la variable cible, indiquant une possible influence significative sur le statut de bon ou mauvais client. Ces variables semblent refléter des aspects importants de la stabilité financière et de l'engagement envers la banque.
*   **Autres Variables Pertinentes :** L'ancienneté et la moyenne des mouvements financiers montrent également des corrélations notables avec le type de client, suggérant que la durée de la relation avec la banque et l'activité du compte peuvent jouer un rôle dans l'évaluation de la fiabilité du client.

<span style="color: red;"><strong>Pourquoi ne pas Supprimer des Xi​ Fortement Corrélées ? :</strong></span>

*   **Pertinence des Variables :** Même si certaines variables sont corrélées, chacune peut apporter des informations uniques et précieuses qui contribuent à une prédiction plus riche et nuancée. Par exemple, bien que "Domiciliation du salaire" et "Moyenne encours" puissent partager certaines informations sur la stabilité financière, elles capturent différents aspects du comportement financier du client.
*   **Complexité vs Performance :** Dans le cadre d'une régression logistique appliquée à des variables catégorielles, la présence de variables corrélées n'entraîne pas nécessairement les mêmes problèmes de multicollinéarité que dans une régression linéaire. La suppression de variables corrélées sans une évaluation approfondie pourrait éliminer des informations importantes, sans améliorer significativement la performance du modèle.
*   **Interprétation et Signification :** Chaque variable catégorielle a une interprétation spécifique et une signification dans le contexte métier. Supprimer des variables uniquement en raison de leur corrélation pourrait réduire la capacité du modèle à fournir des insights actionnables et compréhensibles pour la prise de décision.

<span style="color: red;"><strong>Conclusion sur l'Analyse des Corrélations et Intérêt:</strong></span>

La décision de conserver les variables corrélées est soutenue par la volonté de préserver l'intégralité des informations disponibles et de maintenir une interprétation riche et nuancée des facteurs influençant la fiabilité des clients. Dans notre cas, l'accent est mis sur l'utilisation judicieuse des données disponibles pour construire un modèle de régression logistique qui soit à la fois prédictif et informatif, sans compromettre la complexité nécessaire pour capturer la dynamique du comportement des clients vis-à-vis de l'octroi de crédit.

```python
def highlight_over_50_percent(val):
    """
    Applique un surlignage de fond aux cellules du DataFrame ayant des valeurs > 50.
    
    Args:
    val: La valeur de la cellule du DataFrame.
    
    Returns:
    Un style CSS sous forme de chaîne de caractères à appliquer à la cellule.
    """
    color = 'yellow' if val > 50 else ''
    return f'background-color: {color}'

def analyse_bivariée_complète_avec_decompte(variable, data, order):
    # Tableau croisé pour décompte réel transformé en pourcentage
    crosstab_percentage = pd.crosstab(data[variable], data['Type de client'], normalize='index') * 100
    crosstab_percentage = crosstab_percentage.reindex(order)
    
    # Affichage du tableau avec surlignage pour les pourcentages > 50%
    styled_crosstab = crosstab_percentage.style.applymap(highlight_over_50_percent).format("{:.2f}%")
    display(styled_crosstab)
    
    # Graphique à barres avec étiquettes pour les pourcentages
    colors = ['green', 'orange']  # Assumption: the order of colors matches the order of 'Type de client'
    ax = crosstab_percentage.plot(kind='bar', stacked=True, color=colors, figsize=(10, 6))
    plt.title(f'Distribution de "Type de Client" par {variable}')
    plt.xlabel(variable)
    plt.ylabel('Pourcentage')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Type de Client', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    
    # Ajout des étiquettes de pourcentage dans les barres
    for rect in ax.patches:
        height = rect.get_height()
        if height > 0:  # Éviter l'affichage des étiquettes pour les valeurs nulles
            ax.text(rect.get_x() + rect.get_width() / 2, rect.get_y() + height / 2, f"{height:.1f}%", 
                    ha='center', va='center', color='white', size=8)

    plt.show()
    
# Appliquons cette fonction ajustée pour chaque variable catégorielle en prenant en compte l'ordre correct et les regroupements
for variable, order in orders_corrected.items():
    analyse_bivariée_complète_avec_decompte(variable, data, order)
```

### 6. Division des Données et Entraînement du Modèle <a id="6-division-des-données-et-entraînement-du-modèle"></a>

**Introduction**  
Cette étape marque le cœur de notre processus de modélisation. Elle comprend la division de notre jeu de données en ensembles d'entraînement et de test, suivie de l'entraînement du modèle sur l'ensemble d'entraînement. Cette phase est cruciale car elle prépare notre modèle à faire des prédictions fiables sur des données non vues.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML

# Définition de la fonction pour recalculer les métriques
def recalculate_metrics(y_true, y_pred, y_pred_proba):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_pred_proba)
    
    # Calcul de la spécificité
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp)
    
    return {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'Specificity': specificity, 'ROC AUC': roc_auc}

# Définition de X et y

# Transformation de 'y' en valeurs numériques
y_num = y.map({'Mauvais client': 0, 'Bon client': 1})

# Division en données d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y_num, test_size=0.2, random_state=42, stratify=y_num)

# Entraînement du modèle
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Évaluation du modèle
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
y_train_pred_proba = model.predict_proba(X_train)[:, 1]
y_test_pred_proba = model.predict_proba(X_test)[:, 1]

# Calcul des métriques
train_metrics = recalculate_metrics(y_train, y_train_pred, y_train_pred_proba)
test_metrics = recalculate_metrics(y_test, y_test_pred, y_test_pred_proba)

# Affichage des métriques sous forme de tableau HTML avec un titre explicatif
print("Résumé des Métriques de Performance du Modèle de Classification :")
metrics_summary = pd.DataFrame({
    "Métrique": ["Accuracy (Exactitude)", "Precision (Précision)", "Recall (Rappel)", "Specificity (Spécificité)", "AUC (Aire sous la courbe ROC)"],
    "Entraînement": list(train_metrics.values()),
    "Test": list(test_metrics.values())
})
display(HTML(metrics_summary.to_html(index=False)))

# Affichage des matrices de confusion avec titres explicatifs sur trois lignes
def plot_confusion_matrix_french_with_all_metrics(y_true, y_pred, y_pred_proba, dataset="Ensemble d'Entraînement ou de Test"):
    accuracy, precision, recall, specificity, _ = recalculate_metrics(y_true, y_pred, y_pred_proba).values()
    title = (f"Matrice de Confusion pour l'{dataset}\n"
             f"Exactitude (Accuracy): {accuracy:.2f}, Précision (Precision): {precision:.2f},\n"
             f"Rappel (Recall): {recall:.2f}, Spécificité (Specificity): {specificity:.2f}")
    
    conf_mat = confusion_matrix(y_true, y_pred)
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Positif', 'Négatif'], yticklabels=['Positif', 'Négatif'])
    plt.title(title, pad=20)
    plt.xlabel('Valeurs Prédites')
    plt.ylabel('Valeurs Réelles')

# Affichage des matrices de confusion avec titres explicatifs sur trois lignes
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plot_confusion_matrix_french_with_all_metrics(y_train, y_train_pred, y_train_pred_proba, "Ensemble d'Entraînement")

plt.subplot(1, 2, 2)
plot_confusion_matrix_french_with_all_metrics(y_test, y_test_pred, y_test_pred_proba, "Ensemble de Test")

plt.tight_layout(pad=2.0)
plt.show()
```

Dans cette section, nous utilisons la courbe ROC pour évaluer visuellement la capacité de notre modèle de régression logistique à distinguer entre les "Bons clients" et les "Mauvais clients".

1.  **Courbe ROC** : La courbe trace le taux de vrais positifs (sensibilité) contre le taux de faux positifs (1 - spécificité) à différents seuils de décision. Plus la courbe se rapproche du coin supérieur gauche, meilleure est la performance du modèle.
    
2.  **Aire Sous la Courbe (AUC)** : L'AUC résume la performance du modèle en un seul nombre. Une valeur de 1.0 représente une capacité parfaite à distinguer les classes, tandis qu'une valeur de 0.5 indique une performance équivalente à un choix aléatoire.
    

En résumé, l'utilisation conjointe des courbes ROC et des scores AUC fournit un aperçu visuel et quantitatif de la performance du modèle, soulignant son efficacité à distinguer entre les deux classes de clients. La courbe ROC nous aide également à comprendre la justesse et la robustesse de notre modèle dans la prédiction du type de client, en fournissant une mesure visuelle et quantitative de sa performance.

**Introduction à l'Affinement du Modèle de Scoring**

Nous nous apprêtons à déployer une méthodologie rigoureuse de sélection de variables pour notre modèle de scoring de crédit. L'élément central de cette démarche est l'utilisation des p-valeurs, ces balises statistiques qui déterminent l'importance de chaque variable dans la prédiction du risque client.

Les variables présentant une influence statistiquement non significative seront progressivement écartées, affinant ainsi notre modèle pour qu'il reflète avec précision les facteurs véritablement déterminants pour le comportement de remboursement de nos clients.

L'enjeu ici est double : gagner en précision prédictive tout en affûtant notre compréhension des dynamiques de risque, aboutissant à un outil de décision plus robuste et stratégique dans l'évaluation du risque de crédit.

```python
import pandas as pd
import statsmodels.api as sm

# Utilisation de `drop_first=True` pour éviter la multicollinéarité
data_2 = pd.get_dummies(data.drop(["Identifiant Client", "Score 1", "Score 2", "UniqueID", "Type de client"], axis=1), drop_first=True)

y = data["Type de client"].replace(["Mauvais client", "Bon client"], [0, 1])

# Conversion des variables booléennes en entiers
data_2 = data_2.astype(int)

# Ajout d'une constante (intercept) au modèle
X_filtre = sm.add_constant(data_2)

# Initialisation d'une liste pour stocker les variables supprimées avec leurs P-valeurs
variables_supprimees = []

# Boucle pour retirer les variables avec les plus grandes P-valeurs supérieures à 0.05 une par une
while True:
    model_filtre = sm.Logit(y, X_filtre).fit(disp=0)  # disp=0 pour éviter l'impression de l'output pendant la convergence
    
    # Obtenir les P-valeurs et retirer 'const' de l'analyse
    p_values = model_filtre.pvalues.drop('const', errors='ignore')
    
    # Trouver la variable avec la plus grande P-valeur supérieure à 0.05
    max_p_value = p_values.max()
    
    if max_p_value > 0.05:
        # Retirer la variable associée à la plus grande P-valeur
        variable_a_retirer = p_values.idxmax()
        variables_supprimees.append((variable_a_retirer, max_p_value))
        X_filtre = X_filtre.drop(columns=[variable_a_retirer])
    else:
        break  # Sortir de la boucle si aucune P-valeur n'est supérieure à 0.05

# Afficher le résumé du modèle final
print("Modèle après sélection de variables :")
print(model_filtre.summary())

# Extraire les coefficients et autres statistiques
output_table = pd.DataFrame({
    'Variable': model_filtre.params.index,
    'Coefficient': model_filtre.params.values,
    'Erreur standard': model_filtre.bse.values,
    'P-Valeur': model_filtre.pvalues.values
})

# Afficher le tableau des résultats
print(output_table)

# Afficher la liste des variables supprimées avec leurs P-valeurs
print("Liste des variables supprimées avec leurs P-valeurs :")
for var, pval in variables_supprimees:
    print(f"Variable : {var}, P-valeur : {pval}")
```

3. **Décisions Basées sur des Règles** : Les arbres de décision fonctionnent en créant des règles de décision basées sur les caractéristiques des données. Cela ressemble à la manière dont les décisions sont souvent prises dans les contextes commerciaux, rendant le modèle pertinent pour des applications pratiques.
